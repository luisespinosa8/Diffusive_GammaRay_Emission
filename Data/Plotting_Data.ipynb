# Plots of diffusive gamma-ray emission data

## TIBET

### Inner Galaxy

```python
# TIBET region 25<l<100 and |b|<5. Energy in TeV and dN/dE*E^2.7 (GeV^1.7 cm^-2 s^-1 sr^-1) 
data_TIBET_innerGalaxy = np.genfromtxt('TIBET/TIBET_innerGalaxy.txt')
E_TIBET_innerGalaxy = data_TIBET_innerGalaxy[:,0] * 1e3
E_max_TIBET_innerGalaxy = data_TIBET_innerGalaxy[:,1] * 1e3
E_min_TIBET_innerGalaxy = data_TIBET_innerGalaxy[:,2] * 1e3
phi_TIBET_innerGalaxy = data_TIBET_innerGalaxy[:,3]
phi_max_TIBET_innerGalaxy = data_TIBET_innerGalaxy[:,4]
phi_min_TIBET_innerGalaxy = data_TIBET_innerGalaxy[:,5]
```

```python
# Fermi region 25<l<100 and |b|<5. Energy in MeV and dN/dE*E (MeV^-1 cm^-2 s^-1 sr^-1) 
E_Fermi_innerGalaxy_TIBET = np.genfromtxt('TIBET/Fermi/E_center_inner.txt') * 1e-3
E_max_Fermi_innerGalaxy_TIBET = np.genfromtxt('TIBET/Fermi/xerr_up_inner.txt') * 1e-3
E_min_Fermi_innerGalaxy_TIBET = np.genfromtxt('TIBET/Fermi/xerr_do_inner.txt') * 1e-3
phi_Fermi_innerGalaxy_TIBET = np.genfromtxt('TIBET/Fermi/flux_inner.txt') / 1e-3
phi_max_Fermi_innerGalaxy_TIBET = np.genfromtxt('TIBET/Fermi/yerr_inner.txt') / (E_Fermi_innerGalaxy_TIBET*1e3)**2
phi_min_Fermi_innerGalaxy_TIBET = np.genfromtxt('TIBET/Fermi/yerr_inner.txt') / (E_Fermi_innerGalaxy_TIBET*1e3)**2
phi_max_Fermi_innerGalaxy_TIBET = phi_max_Fermi_innerGalaxy_TIBET / 1e-3
phi_min_Fermi_innerGalaxy_TIBET = phi_min_Fermi_innerGalaxy_TIBET / 1e-3
```

```python
plt.figure(figsize=(10, 5))
plt.errorbar(E_TIBET_innerGalaxy, phi_TIBET_innerGalaxy, yerr=(phi_min_TIBET_innerGalaxy, phi_max_TIBET_innerGalaxy), fmt='o', color='blue', label='TIBET')
plt.errorbar(E_Fermi_innerGalaxy_TIBET, phi_Fermi_innerGalaxy_TIBET * E_Fermi_innerGalaxy_TIBET**alpha, yerr=(phi_min_Fermi_innerGalaxy_TIBET, phi_max_Fermi_innerGalaxy_TIBET)*E_Fermi_innerGalaxy_TIBET**alpha, xerr=(E_min_Fermi_innerGalaxy_TIBET, E_max_Fermi_innerGalaxy_TIBET), fmt='o', color='green', label='Fermi')
plt.loglog()
plt.xlim(1e0, 1e6)
plt.ylim(1e-5, 1e-3)
plt.ylabel(r'$E^{2.7}\phi(E)$ [GeV$^{1.7}$ (cm$^2$ s sr)$^{-1}$]', fontsize=17)
plt.xlabel(r'$E_\gamma$ [GeV]', fontsize=17)
plt.legend(ncol=1, fontsize=15, loc='upper left')
plt.title('Diffusive gamma-ray spectrum in $25<l<100$ and $|b|<5$', fontsize=17)
```
