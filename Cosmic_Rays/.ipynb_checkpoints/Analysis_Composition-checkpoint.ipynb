## My model

```python
# Model for proton
def Phi_proton(E):
    K = 0.0454255
    E0 = 0.1
    alpha1 = 2.805; alpha2 = 2.56; alpha3 = 2.814; alpha4 = 1.335; alpha5 = 4.025
    omega1 = 0.28; omega2 = 0.163; omega3 = 0.648; omega4 = 0.941
    Eb1 = 0.670; Eb2 = 16; Eb3 = 0.799; Eb4 = 3.777
    delta1 = alpha2 - alpha1
    delta2 = alpha3 - alpha2
    delta3 = alpha4 - alpha3
    delta4 = alpha5 - alpha4
    phi = K * ((E/1e3) / E0)**(-alpha3) * (1 + ((E/1e3) / Eb1)**(1 / omega1))**(-delta1 * omega1) * (1 + ((E/1e3) / Eb2)**(1 / omega2))**(-delta2 * omega2) * (1 + ((E/1e6) / Eb3)**(1 / omega3))**(-delta3 * omega3) * (1 + ((E/1e6) / Eb4)**(1 / omega4))**(-delta4 * omega4)
    return phi

# Model for He
def Phi_He(E):
    n = 4
    K = 0.0332187 
    E0 = 0.1
    alpha1 = 2.702; alpha2 = 2.464; alpha3 = 2.733; alpha4 = 2.3; alpha5 = 4.22
    omega1 = 0.272; omega2 = 0.056; omega3 = 0.1; omega4 = 0.701
    Eb1 = 1.207; Eb2 = 32.014; Eb3 = 0.526; Eb4 = 6.168
    delta1 = alpha2 - alpha1
    delta2 = alpha3 - alpha2
    delta3 = alpha4 - alpha3
    delta4 = alpha5 - alpha4
    phi = K * ((E/1e3) / E0)**(-alpha3) * (1 + ((E/1e3) / Eb1)**(1 / omega1))**(-delta1 * omega1) * (1 + ((E/1e3) / Eb2)**(1 / omega2))**(-delta2 * omega2) * (1 + ((E/1e6) / Eb3)**(1 / omega3))**(-delta3 * omega3) * (1 + ((E/1e6) / Eb4)**(1 / omega4))**(-delta4 * omega4)
    return phi

def Phi_shape(E, Z):
    E0 = 0.1
    alpha1 = 2.702; alpha2 = 2.464; alpha3 = 2.733; alpha4 = 2.3; alpha5 = 4.22
    omega1 = 0.272; omega2 = 0.056; omega3 = 0.1; omega4 = 0.701
    Eb1 = 1.207 * (Z/2); Eb2 = 32.014 * (Z/2); Eb3 = 0.526 * (Z/2); Eb4 = 6.168 * (Z/2)
    delta1 = alpha2 - alpha1
    delta2 = alpha3 - alpha2
    delta3 = alpha4 - alpha3
    delta4 = alpha5 - alpha4
    phi = ((E/1e3) / E0)**(-alpha3) * (1 + ((E/1e3) / Eb1)**(1 / omega1))**(-delta1 * omega1) * (1 + ((E/1e3) / Eb2)**(1 / omega2))**(-delta2 * omega2) * (1 + ((E/1e6) / Eb3)**(1 / omega3))**(-delta3 * omega3) * (1 + ((E/1e6) / Eb4)**(1 / omega4))**(-delta4 * omega4)
    return phi

# Model for heavy nuclei
def Phi_heavy(E, A, Z):
    #KA = 1 * (0.0454255 + (0.0332187 * 2**2.702)) * (2**2.702 * A**-2.702)
    KA = 0.0454255 + 0.0332187#(0.0454255 + (0.0332187 * 2**2.702)) / Z**2.702
    return KA * Phi_shape(E, Z)
```

```python
lnA_C = ( (np.log(1) * Phi_proton(E)) + (np.log(4) * Phi_He(E)) + (np.log(12) * Phi_heavy(E, 12, 6)) ) / (Phi_proton(E) + Phi_He(E) + Phi_heavy(E, 12, 6))
lnA_Fe = ( (np.log(1) * Phi_proton(E)) + (np.log(4) * Phi_He(E)) + (np.log(56) * Phi_heavy(E, 56, 26)) ) / (Phi_proton(E) + Phi_He(E) + Phi_heavy(E, 56, 26))
```

```python
plt.plot(E, Phi_proton(E)*E**2.7, label='p')
plt.plot(E, Phi_He(E)*E**2.7, label='He')
plt.plot(E, Phi_heavy(E, 12, 6)*E**2.7, label='C')
plt.plot(E, Phi_heavy(E, 56, 26)*E**2.7, label='Fe')

plt.loglog()
plt.xlim(1e1, 1e8)
#plt.ylim(1e3, 1e5)
plt.legend()
```

## Dembinski model

```python
data = np.genfromtxt('CRdatadriven.txt')
A = np.array((0, 1, 4, 7, 9, 10, 12, 14, 16, 19, 20, 23, 24, 27, 28, 31, 32, 35, 40, 39, 40, 45, 48, 51, 52, 55, 56, 59, 59))
sum_phi = np.zeros(len(E))
sum_phi_lnA = np.zeros(len(E))
for i in range (1,29):
    phi_interpolated = scipy.interpolate.PchipInterpolator(data[:,0] * A[i], (data[:,i] * 1e4) / (A[i]**2))
    sum_phi += phi_interpolated(E)
    sum_phi_lnA += phi_interpolated(E) * np.log(A[i])
lnA_Dembinski = sum_phi_lnA / sum_phi
```

## Prevotat model

```python
# Values chosen for DAMPE analysis
def Intensity(Z, N, E0, E1, gamma, E, n):
    E0 = Z * E0 
    E1 = Z * E1
    I = (N / E0**2.5) * (E / E0)**(-gamma) * np.exp( - (E0 / E) - (E / E1))
    return I

# Protons
def I_p(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E):
    I1 = Intensity(1, N1, E01, E11, gamma1, E, 1)
    I2 = Intensity(1, N2, E02, E12, gamma2, E, 1)
    I3 = Intensity(1, N3, E03, E13, gamma3, E, 1)
    I4 = Intensity(1, N4, E04, E14, gamma4, E, 1)
    I_total = I1 + I2 + I3 + I4
    return I_total

N1 = 2.03e4; N2 = 3.67e3; N3 = 1.13e3; N4 = 2.55e3
E01 = 6.21e0; E02 = 2.20e1; E03 = 4.41e3; E04 = 7.90e5
E11 = 4.06e2; E12 = 2.81e4; E13 = 3.08e5; E14 = 1.90e7
gamma1 = 3.25; gamma2 = 2.59; gamma3 = 2.42; gamma4 = 3.02
intensity_p = I_p(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E)

# Helium
def I_He(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E):
    I1 = Intensity(2, N1, E01, E11, gamma1, E, 4)
    I2 = Intensity(2, N2, E02, E12, gamma2, E, 4)
    I3 = Intensity(2, N3, E03, E13, gamma3, E, 4)
    I4 = Intensity(2, N4, E04, E14, gamma4, E, 4)
    I_total = I1 + I2 + I3 + I4
    return I_total

N1 = 1.06e4; N2 = 2.73e3; N3 = 1.38e3; N4 = 1.73e3
E01 = 6.21e0; E02 = 2.20e1; E03 = 4.41e3; E04 = 7.90e5
E11 = 4.06e2; E12 = 2.81e4; E13 = 3.08e5; E14 = 1.90e7
gamma1 = 3.25; gamma2 = 2.59; gamma3 = 2.42; gamma4 = 3.02
intensity_He = I_He(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E)

# CNO
def I_CNO(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E):
    I1 = Intensity(7, N1, E01, E11, gamma1, E, 14)
    I2 = Intensity(7, N2, E02, E12, gamma2, E, 14)
    I3 = Intensity(7, N3, E03, E13, gamma3, E, 14)
    I4 = Intensity(7, N4, E04, E14, gamma4, E, 14)
    I_total = I1 + I2 + I3 + I4
    return I_total

N1 = 7.20e3; N2 = 2.10e3; N3 = 1.23e3; N4 = 7.01e2
E01 = 6.21e0; E02 = 2.20e1; E03 = 4.41e3; E04 = 7.90e5
E11 = 4.06e2; E12 = 2.81e4; E13 = 3.08e5; E14 = 1.90e7
gamma1 = 3.25; gamma2 = 2.59; gamma3 = 2.42; gamma4 = 3.02
intensity_CNO = I_CNO(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E)

# Iron
def I_Fe(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E):
    I1 = Intensity(26, N1, E01, E11, gamma1, E, 56)
    I2 = Intensity(26, N2, E02, E12, gamma2, E, 56)
    I3 = Intensity(26, N3, E03, E13, gamma3, E, 56)
    I4 = Intensity(26, N4, E04, E14, gamma4, E, 56)
    I_total = I1 + I2 + I3 + I4
    return I_total

N1 = 2.50e3; N2 = 1.03e3; N3 = 1.00e2; N4 = 3.50e2
E01 = 6.21e0; E02 = 2.20e1; E03 = 4.41e3; E04 = 7.90e5
E11 = 4.06e2; E12 = 2.81e4; E13 = 3.08e5; E14 = 1.90e7
gamma1 = 3.25; gamma2 = 2.59; gamma3 = 2.42; gamma4 = 3.02
intensity_Fe = I_Fe(N1, E01, E11, gamma1, N2, E02, E12, gamma2, N3, E03, E13, gamma3, N4, E04, E14, gamma4, E)
```

```python
lnA_Prevotat = ( (np.log(1) * intensity_p) + ( np.log(4) * intensity_He) + (np.log(14) * intensity_CNO) + (np.log(56) * intensity_Fe) ) / (intensity_p + intensity_He + intensity_CNO + intensity_Fe)
```

## Plotting $< \ln A>$ 

```python
# Auger data
data_Auger = np.genfromtxt('Composition_Data/AugerICRC2019_lnA_Moments_EPOS-LHC.txt')
energy_Auger = 10**data_Auger[:,0] / 1e6
lnA_Auger = data_Auger[:,1]

# LHAASO
data_LHAASO = np.genfromtxt('Composition_Data/LHAASO_EPOS-LHC_lnA_energy.txt')
energy_LHAASO = data_LHAASO[:,0]
lnA_LHAASO = data_LHAASO[:,1]

plt.figure(figsize=(15, 7))
# Models
plt.plot(E, lnA_Fe, color='blue', label='My model (w/ Fe)')
plt.plot(E, lnA_C, color='gold', label='My model (w/ C)')
plt.plot(E, lnA_Dembinski, color='red', label='Dembinski model')
plt.plot(E, lnA_Prevotat, color='green', label='Prevotat model')
# Data
plt.errorbar(energy_LHAASO, lnA_LHAASO, fmt='s', color='black', label='LHAASO')
plt.errorbar(energy_Auger, lnA_Auger, fmt='o', color='black', label='Auger')
plt.loglog()
plt.xlim(1e2, 1e14)
plt.ylim(5e-1, 1e1)
plt.axhline(y = np.log(4), color='black', linestyle='--', label='He')
plt.axhline(y = np.log(14), color='black', linestyle='-.', label='CNO')
plt.axhline(y = np.log(56), color='black', linestyle=':', label='Fe')
plt.xlabel('E [GeV]', fontsize=17)
plt.ylabel(r'$< \ln A>$', fontsize=17)
plt.legend(ncol=3, fontsize=14, loc='upper left')
plt.title('Hadronic model: EPOS-LHC', fontsize=17)
```
