# Diffusive gamma-ray emission measured by Fermi-LAT

```python
from pylab import *
```

```python
E_Fermi = np.genfromtxt('E_center.txt')
flux_Fermi = np.genfromtxt('flux.txt')
error_y = np.genfromtxt('yerr.txt') / E_Fermi**2
error_x_up = np.genfromtxt('xerr_up.txt') 
error_x_down = np.genfromtxt('xerr_do.txt') 
```

```python
# MeV to GeV
E_Fermi = E_Fermi * 1e-3
# (MeV cm^2 s sr)^-1 to (GeV cm^2 s sr)^-1
flux_Fermi = flux_Fermi / 1e-3
error_x_up = error_x_up * 1e-3
error_x_down = error_x_down * 1e-3
error_y = error_y / 1e-3
```

```python
alpha = 2.7

plt.figure(figsize=(10, 7))
plt.errorbar(E_Fermi, flux_Fermi * E_Fermi**alpha, yerr=error_y * E_Fermi**alpha, xerr=(error_x_down, error_x_up), fmt='o', color='red')
plt.loglog()
plt.xlim(1e-1, 1e3)
plt.ylim(1e-5, 5e-4)
plt.xlabel('E [GeV]', fontsize=17)
plt.ylabel(r'$E^{2.7}\phi(E)$ [GeV$^{1.7}$ (m$^2$ s sr)$^{-1}$]', fontsize=17)
```
